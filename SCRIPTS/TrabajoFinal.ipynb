```python
SB_Agrupado = SB.groupby(["fuente", "Year","Month"])[["fuente", "Year","Month","Active_energy","Reactive_energy","Voltaje_FA","Voltaje_FC"]].agg({
    'fuente':'first',
    'Year':'first',
    'Month':'first',
    "Active_energy":'sum',
    "Reactive_energy":'sum',
    "Voltaje_FA":'sum',
    "Voltaje_FC":'sum'
    }
)
SB_Agrupado
```

```python
SB_Agrupado = SB_Agrupado[~((SB_Agrupado['Year']==2023) & (SB_Agrupado['Month']==4))]
```

```python
SB_Agrupado=SB_Agrupado.reset_index(drop=True)
```

```python
def identify_outliers(group, threshold=3):
    mean = group['Active_energy'].mean()
    std_dev = group['Active_energy'].std()
    group['Outlier_Active_energy'] = group['Active_energy'].apply(lambda x: 1 if np.abs((x - mean) / std_dev) > threshold else 0)
    
    mean = group['Reactive_energy'].mean()
    std_dev = group['Reactive_energy'].std()
    group['Outlier_Reactive_energy'] = group['Reactive_energy'].apply(lambda x: 1 if np.abs((x - mean) / std_dev) > threshold else 0)
    
    mean = group['Voltaje_FA'].mean()
    std_dev = group['Voltaje_FA'].std()
    group['Outlier_Voltaje_FA'] = group['Voltaje_FA'].apply(lambda x: 1 if np.abs((x - mean) / std_dev) > threshold else 0)
    
    mean = group['Voltaje_FC'].mean()
    std_dev = group['Voltaje_FC'].std()
    group['Outlier_Voltaje_FC'] = group['Voltaje_FC'].apply(lambda x: 1 if np.abs((x - mean) / std_dev) > threshold else 0)
    
    return group

# Apply the function to each group
outlier_df = SB_Agrupado.groupby('fuente').apply(identify_outliers)
```

```python
Anomalias_1 = outlier_df.drop(['Active_energy','Reactive_energy','Voltaje_FA',	'Voltaje_FC',	'Outlier_Active_energy',	'Outlier_Reactive_energy',	'Outlier_Voltaje_FA',	'Outlier_Voltaje_FC'], axis=1)
Anomalias_1.loc[Anomalias_1.Anomalia>0, 'Anomalia']=1
Anomalias_1
```

```python
datos = outlier_df.drop([ 'Outlier_Active_energy','Outlier_Reactive_energy', 'Outlier_Voltaje_FA', 'Outlier_Voltaje_FC',
       'Anomalia'], axis=1)
datos
```

```python
# Configurando el Isolation Forest
clf = IsolationForest(n_estimators=100, max_samples='auto', contamination=float(.12), \
                      max_features=1.0, bootstrap=False, n_jobs=-1, random_state=42, verbose=0)

# Entrenar el Isolation Forest
clf.fit(datos)

# Predecir si una observaciÃ³n es una anomalÃ­a. -1 indica anomalÃ­a, 1 indica normal.
preds = clf.predict(datos)

# AÃ±adir las predicciones al DataFrame
datos['anomaly'] = preds
```

```python
anomalies = datos[datos['anomaly'] == -1]
print(anomalies)
```

```python
Anomalias_2 = anomalies.drop(['Active_energy','Reactive_energy','Voltaje_FA',	'Voltaje_FC'], axis=1)
```

```python
Validacion=Anomalias_1.merge(Anomalias_2,how='outer')
Validacion['anomaly'] = Validacion['anomaly'].fillna(0)
Validacion
```

```python
Validacion['resultado']=Validacion['Anomalia']+Validacion['anomaly']
Validacion.groupby(['fuente','Year', 'Month'])[['Anomalia',	'anomaly','resultado']].agg("sum").sort_values(by ='resultado', ascending=False)
Validacion.head()
```
